# **Questões a serrem respondidas dos problemas em questão**

A desigualdade social e de renda, pode impactar diversos fatores e setores, como o de emissão de carbono, da indústria, e até o digital. Com a desigualdade, a concentração de renda em apenas uma parcela da população pode desencadear problemas socioeconômicos, dificuldade de acesso a novas tecnologias, e influenciar no leque de oportunidades que os gêneros podem ter, causando uma disparidade de oportunidades e renda dentro de uma população e entre gêneros.

Com a desigualdade, com a concentração de renda em uma certa parcela desencadeia que a parcela sem o acúmulo de renda busca por recursos que acabam tendo uma maior emissão de carbono, podendo afetar seriamente a saúde do indivíduo. 

Todos esses fatores podem vir diretamente de más decisões de políticas governamentais que acabam influenciando a concentração de renda em uma pequena parte da população.

##1) Qual relação entre as políticas do governo e a desigualdade social?

A relação existente entre as políticas de governo e a desigualdade social, se deve ao fato de que, as políticas adotadas pelos governos podem prejudicar a igualdade social. Na América Latina, podemos observar que quanto maior a desigualdade, menor é o número de impostos cobrados, enquanto que na África, quanto maior a desigualdade, maior os impostos cobrados. Na Europa já não conseguimos fazer essa relação.

```python
import requests
from io import BytesIO
url = 'https://github.com/andresbonito/Ciencia-de-Dados---IMT-Maua/raw/main/Questão1.xlsx'
response = requests.get(url)
content = response.content

wid = pd.read_excel(BytesIO(content))
# Importe os dados do World Inequality Database

# Provide the path to your XLS file

# Read the XLS file into a pandas DataFrame
#wid = pd.read_excel(path)


# Responda à pergunta 1: Qual é a relação entre as políticas governamentais e a desigualdade social?

# Crie um novo dataframe que mostre a relação entre os gastos do governo em programas sociais e o coeficiente de Gini
social_spending = wid[["country", "year", "gini Europa", "gini América Latina", "gini África", "Number of tax returns"]]

# Trace a relação entre gastos sociais e coeficiente de Gini
plt.scatter(social_spending["Number of tax returns"], social_spending["gini Europa"])
plt.xlabel("Number of tax returns")
plt.ylabel("gini Europa")
plt.title("Relação entre o coeficiente de gini e a cobrança de impostos")
plt.show()

# Trace a relação entre gastos sociais e coeficiente de Gini
plt.scatter(social_spending["Number of tax returns"], social_spending["gini América Latina"])
plt.xlabel("Number of tax returns")
plt.ylabel("gini América Latina")
plt.title("Relação entre o coeficiente de gini e a cobrança de impostos")
plt.show()

# Trace a relação entre gastos sociais e coeficiente de Gini
plt.scatter(social_spending["Number of tax returns"], social_spending["gini África"])
plt.xlabel("Number of tax returns")
plt.ylabel("gini África")
plt.title("Relação entre o coeficiente de gini e a cobrança de impostos")
plt.show()
```

```python
social_spending2 = social_spending.tail(42)
```

```python
social_spending2['gini World'] = (social_spending2['gini Europa'] + social_spending2['gini África'] + social_spending2['gini América Latina']) / 3
social_spending2
```

```python
# Trace a relação entre gastos sociais e coeficiente de Gini
plt.scatter(social_spending2["Number of tax returns"], social_spending2["gini World"])
plt.xlabel("Number of tax returns")
plt.ylabel("gini World")
plt.title("Relação entre o coeficiente de gini e a cobrança de impostos")
plt.show()
```

## 2) Como a desigualdade afeta a emissão de carbono no meio ambiente?

A desigualdade pode afetar a emissão de carbono por ter diferentes padrões de consumo, já que em um país com renda mal distribuída, pode afetar as regiões mais pobres, por não terem acesso a energias limpas. Então, quanto mais desequilibrado for a distribuição de renda, maior será a emissão de carbono.

```python
import requests
from io import BytesIO

url = 'https://github.com/andresbonito/Ciencia-de-Dados---IMT-Maua/raw/main/Questão2.xls'
response = requests.get(url)
content = response.content

wid = pd.read_excel(BytesIO(content))
# Read the XLS file into a pandas DataFrame

# Responda à pergunta 2: Como a desigualdade afeta as emissões de carbono no meio ambiente?

# Crie um novo dataframe que mostre a relação entre o coeficiente de Gini e as emissões de carbono per capita
carbon_emissions = wid[["country", "year", "gini Europa", "gini América Latina", "gini Africa", "gini América do Norte", "carbon_emissions_per_capita"]]

# Trace a relação entre o coeficiente de Gini e as emissões de carbono per capita
plt.scatter(carbon_emissions["gini Europa"], carbon_emissions["carbon_emissions_per_capita"])
plt.xlabel("gini Europa")
plt.ylabel("Carbon emissions per capita")
plt.title("Relação entre o coeficiente de gini e as emissões de carbono")
plt.show()

# Trace a relação entre o coeficiente de Gini e as emissões de carbono per capita
plt.scatter(carbon_emissions["gini América Latina"], carbon_emissions["carbon_emissions_per_capita"])
plt.xlabel("gini América Latina")
plt.ylabel("Carbon emissions per capita")
plt.title("Relação entre o coeficiente de gini e as emissões de carbono")
plt.show()

# Trace a relação entre o coeficiente de Gini e as emissões de carbono per capita
plt.scatter(carbon_emissions["gini Africa"], carbon_emissions["carbon_emissions_per_capita"])
plt.xlabel("gini Africa")
plt.ylabel("Carbon emissions per capita")
plt.title("Relação entre o coeficiente de gini e as emissões de carbono")
plt.show()

# Trace a relação entre o coeficiente de Gini e as emissões de carbono per capita
plt.scatter(carbon_emissions["gini América do Norte"], carbon_emissions["carbon_emissions_per_capita"])
plt.xlabel("gini América do Norte")
plt.ylabel("Carbon emissions per capita")
plt.title("Relação entre o coeficiente de gini e as emissões de carbono")
plt.show()
```

```python
carbon_emissions
```

```python
carbon_emissions['gini World'] = (carbon_emissions['gini Africa'] + carbon_emissions['gini América Latina'] + carbon_emissions['gini América do Norte'] + carbon_emissions['gini Europa']) / 4
carbon_emissions
```

```python
# Trace a relação entre gastos sociais e coeficiente de Gini
plt.scatter(carbon_emissions['gini World'], carbon_emissions['carbon_emissions_per_capita'])
plt.xlabel("carbon_emissions_per_capita")
plt.ylabel("gini World")
plt.title("Relação entre o coeficiente de gini e a emissão de carbono")
plt.show()
```

## 4)  Como a desigualdade de gênero e a desigualdade de riqueza se relacionam?

A desigualdade de gênero e de renda se relacionam proporcionalmente, já que quanto maior o índice de gini, mais aumenta a desigualdade de gênero, pois essa desigualdade pode gerar uma diferença salarial entre gêneros muito alta, o que afeta a seguir a renda de cada gênero, e assim, diminuindo as oportunidades de mercado que o gênero prejudicado tenha.

```python
import requests
from io import BytesIO

url = 'https://raw.githubusercontent.com/andresbonito/Ciencia-de-Dados---IMT-Maua/main/Quest%C3%A3o%203.xls'
wid = pd.read_excel(url)
# Answer question 3: Como a desigualdade de gênero e a desigualdade de riqueza se relacionam?

# Responda à pergunta 4: Impacto da Desigualdade de Gênero.
gender = wid[["country", "year", "gini Europa", "gini América Latina", "gini África", "gini América do Norte", "Desigualdade de Gênero"]]

# Trace a relação entre o coeficiente de Gini e a parcela de emprego no setor digital
plt.scatter(gender["gini Europa"], gender["Desigualdade de Gênero"])
plt.xlabel("gini Europa")
plt.ylabel("Desigualdade de Gênero")
plt.title("Relação entre o coeficiente de gini e a desigualdade de gênero")
plt.show()

# Trace a relação entre o coeficiente de Gini e a parcela de emprego no setor digital
plt.scatter(gender["gini América Latina"], gender["Desigualdade de Gênero"])
plt.xlabel("gini América Latina")
plt.ylabel("Desigualdade de Gênero")
plt.title("Relação entre o coeficiente de gini e a desigualdade de gênero")
plt.show()

# Trace a relação entre o coeficiente de Gini e a parcela de emprego no setor digital
plt.scatter(gender["gini África"], gender["Desigualdade de Gênero"])
plt.xlabel("gini África")
plt.ylabel("Desigualdade de Gênero")
plt.title("Relação entre o coeficiente de gini e a desigualdade de gênero")
plt.show()

# Trace a relação entre o coeficiente de Gini e a parcela de emprego no setor digital
plt.scatter(gender["gini América do Norte"], gender["Desigualdade de Gênero"])
plt.xlabel("gini América do Norte")
plt.ylabel("Desigualdade de Gênero")
plt.title("Relação entre o coeficiente de gini e a desigualdade de gênero")
plt.show()



```

## 5) A relação entre a desigualdade e a parcela da renda nacional na indústria?

O fator que podemos relacionar a indústria com a desigualdade seria que a concentração de renda em uma certa população, pode diminuir o acesso da parcela mais pobre de ter acesso a industria, além de prejudicar a geração de empregos para essa parcela por fatores socioeconômicos, e assim, mantendo o padrão de apenas uma certa população ter a renda maior

```python
import requests
from io import BytesIO

url = 'https://github.com/andresbonito/Ciencia-de-Dados---IMT-Maua/raw/main/Quest%C3%A3o%204.xlsx'
response = requests.get(url)
content = response.content

wid = pd.read_excel(BytesIO(content))

# Crie um novo dataframe que mostre a relação entre o coeficiente de Gini e a parcela da renda nacional em cada indústria
industry_gini = wid[["country", "year", "gini Europa", "gini América Latina", "gini África", "gini América do Norte", "renda média"]]

# Assuming wid is your DataFrame
renda_media_column = wid["renda média"]

scaler = MinMaxScaler()
renda_media_reshaped = renda_media_column.values.reshape(-1, 1)
renda_media_normalized = scaler.fit_transform(renda_media_reshaped)

wid["renda média"] = renda_media_normalized

# Traçar a relação entre o coeficiente de Gini e a parcela da renda nacional em cada indústria

plt.scatter(industry_gini["gini Europa"], wid["renda média"])  # Updated here
plt.xlabel("gini Europa")
plt.ylabel("Normalized renda média")  # Updated here
plt.title("Relação entre o coeficiente de gini e a renda média")

plt.show()

plt.scatter(industry_gini["gini América Latina"], wid["renda média"])  # Updated here
plt.xlabel("gini América Latina")
plt.ylabel("Normalized renda média")  # Updated here
plt.title("Relação entre o coeficiente de gini e a renda média")

plt.show()

plt.scatter(industry_gini["gini África"], wid["renda média"])  # Updated here
plt.xlabel("gini África")
plt.ylabel("Normalized renda média")  # Updated here
plt.title("Relação entre o coeficiente de gini e a renda média")

plt.show()

plt.scatter(industry_gini["gini América do Norte"], wid["renda média"])  # Updated here
plt.xlabel("gini América do Norte")
plt.ylabel("Normalized renda média")  # Updated here
plt.title("Relação entre o coeficiente de gini e a renda média")

plt.show()




```

## 6) A distribuição de riqueza em diferentes países e como ela se relaciona com a desigualdade de renda?

```python
import requests
from io import BytesIO

url = 'https://github.com/andresbonito/Ciencia-de-Dados---IMT-Maua/raw/main/Quest%C3%A3o%205.xls'
response = requests.get(url)
content = response.content

wid = pd.read_excel(BytesIO(content))

# Responda à pergunta 5: A distribuição da riqueza em diferentes países e como ela se relaciona com a desigualdade de renda?

# Crie um novo dataframe que mostre a relação entre o coeficiente de Gini e o 10% do topo da riqueza
wealth_gini = wid[["country", "year", "gini Europa", "gini América Latina", "gini África", "gini América do Norte", "top_10_share_of_wealth"]]

# Traçar a relação entre o coeficiente de Gini e o 10% mais rico
plt.scatter(wealth_gini["gini Europa"], wealth_gini["top_10_share_of_wealth"])
plt.xlabel("gini Europa")
plt.ylabel("Top 10% share of wealth")
plt.title("Relação entre o coeficiente de gini e a distribuição de riqueza")
plt.show()

# Traçar a relação entre o coeficiente de Gini e o 10% mais rico
plt.scatter(wealth_gini["gini América Latina"], wealth_gini["top_10_share_of_wealth"])
plt.xlabel("gini América Latina")
plt.ylabel("Top 10% share of wealth")
plt.title("Relação entre o coeficiente de gini e a distribuição de riqueza")
plt.show()

# Traçar a relação entre o coeficiente de Gini e o 10% mais rico
plt.scatter(wealth_gini["gini África"], wealth_gini["top_10_share_of_wealth"])
plt.xlabel("gini África")
plt.ylabel("Top 10% share of wealth")
plt.title("Relação entre o coeficiente de gini e a distribuição de riqueza")
plt.show()

# Traçar a relação entre o coeficiente de Gini e o 10% mais rico
plt.scatter(wealth_gini["gini América do Norte"], wealth_gini["top_10_share_of_wealth"])
plt.xlabel("gini América do Norte")
plt.ylabel("Top 10% share of wealth")
plt.title("Relação entre o coeficiente de gini e a distribuição de riqueza")
plt.show()

# Observações

#*A relação entre os gastos do governo com programas sociais e o coeficiente de Gini não é clara. Há alguma evidência de que mais gastos sociais estão associados a níveis mais baixos de desigualdade, mas a relação não é forte.
#*Existe uma relação positiva entre desigualdade e emissões de carbono per capita. Isso sugere que mais desigualdade leva a mais danos ambientais.
#*A transformação digital teve um impacto misto na desigualdade. Em alguns países, o setor digital criou novos empregos e oportunidades, o que ajudou a reduzir a desigualdade. Em outros países, o setor digital levou à perda de empregos e aumentou a desigualdade.
#*As indústrias com maior impacto na desigualdade social variam de país para país. Em alguns países, a agricultura é a indústria mais importante, enquanto em outros países, a indústria ou os serviços são mais importantes.
#*A distribuição da riqueza é altamente desigual em muitos países. O 1% mais rico das famílias geralmente possui uma parcela desproporcional da riqueza, o que pode levar a problemas sociais como pobreza e crime.

#Estes são apenas alguns dos insights que podem ser obtidos com a análise do World Inequality Database. Os dados podem ser usados ​​para responder a uma ampla gama de questões sobre desigualdade e podem ser usados ​​para informar decisões políticas que podem ajudar a reduzir a desigualdade.
```
